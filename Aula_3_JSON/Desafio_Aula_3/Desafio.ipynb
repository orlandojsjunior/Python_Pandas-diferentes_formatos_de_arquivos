# Desafio: normalizando o JSON obtido na API

```python
# Importando as bibliotecas

import pandas as pd
import requests
import json
```

## Obtendo os dados com a API

```python
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

```python
resultado = json.loads(dados_frutas.text)
```

```python
pd.DataFrame(resultado)
```

## Normalizando arquivos JSON

```python
dados_frutas_normalizado = pd.json_normalize(resultado)
```

```python
dados_frutas_normalizado
```

## Escrevendo o arquivo normalizado

```python
dados_frutas_normalizado.to_json('Dados_normalizados.json')
```
